```python
class Handler:
    def __init__(self, successor=None):
        self._successor = successor

    def handle(self, request):
        if self._successor:
            return self._successor.handle(request)
        return None

class AuthenticationHandler(Handler):
    def handle(self, request):
        if request.get('authenticated', False):
            print("User authenticated")
            return super().handle(request)
        else:
            print("Authentication failed")
            return None

class ValidationHandler(Handler):
    def handle(self, request):
        if request.get('valid', False):
            print("Request data is valid")
            return super().handle(request)
        else:
            print("Validation failed")
            return None

class BusinessLogicHandler(Handler):
    def handle(self, request):
        print("Processing business logic")
        return "Business logic processed"

```

```python

# Создаем обработчиков
handler_chain = AuthenticationHandler(
                    ValidationHandler(
                        BusinessLogicHandler()
                    )
                )

# Создаем запрос
request = {'authenticated': True, 'valid': True}

# Передаем запрос в цепочку
result = handler_chain.handle(request)
print(result)
```

```python
from abc import ABC, abstractmethod
import PyPDF2 # type: ignore
from gensim.summarization import summarize # type: ignore

class Handler(ABC):
    def __init__(self):
        self._next_handler = None

    def set_next(self, handler):
        self._next_handler = handler
        return handler  # Возвращает обработчик для возможности "цепочного" вызова

    @abstractmethod
    def handle(self, request):
        if self._next_handler:
            return self._next_handler.handle(request)
        return None


class TextExtractionHandler(Handler):
    def handle(self, request):
        if request == 'extract_text':
            with open(request['file_path'], 'rb') as file:
                reader = PyPDF2.PdfReader(file)
                all_text = ''
                for page_num in range(len(reader.pages)):
                    page = reader.pages[page_num]
                    all_text += page.extract_text()
                clean_text = re.sub(r'(?<=[а-яa-z,-])\s\r?\n(?=[а-яА-Яa-zA-Z])', '', all_text)
                request['text'] = clean_text  # Сохраняем текст в запросе
                request['type'] = 'generate_summary'
            return super().handle(request)
        else:
            return super().handle(request)


class SummaryGenerationHandler(Handler):
    def handle(self, request):
        if 'text' in request and request['type'] == 'generate_summary':
            request['summary'] = summarize(request['text'])
            return super().handle(request)
        else:
            return super().handle(request)
```

```python
# Создаем обработчики
text_extraction_handler = TextExtractionHandler()
summary_generation_handler = SummaryGenerationHandler()

# Настраиваем цепочку
text_extraction_handler.set_next(summary_generation_handler)

# Пример запроса для обработки документа
request = {
    'type': 'extract_text',
    'file_path': '../Data/PDF/text/text2.pdf'
}

# Запуск цепочки обработки
text_extraction_handler.handle(request)

# Результат обработки
print(f"Текст: {request.get('text')}")
print(f"Резюме: {request.get('summary')}")
print(f"Классификация: {request.get('classification')}")
```

```python
import PyPDF2 # type: ignore

def extract_metadata(pdf_path):
    # Open the PDF file
    with open(pdf_path, 'rb') as file:
        # Initialize the PDF reader
        reader = PyPDF2.PdfReader(file)

        # Extract metadata
        metadata = reader.metadata

        # Print metadata
        for key, value in metadata.items():
            print(f"{key}: {value}")

# Example usage
pdf_path = '../Data/PDF/text/text2.pdf'
extract_metadata(pdf_path)
```

```python
def ocr_pdf(file_path):
    # Open the PDF file
    doc = fitz.open(file_path)

    # Iterate through the pages
    for page_num in range(len(doc)):
        page = doc.load_page(page_num)
        images = page.get_images(full=True)

        # Iterate through the images on the page
        for img_index, img in enumerate(images):
            xref = img[0]
            base_image = doc.extract_image(xref)
            image_bytes = base_image["image"]

            # Open the image using PIL
            image = Image.open(io.BytesIO(image_bytes))

            # Perform OCR using pytesseract
            text = pytesseract.image_to_string(image)
            print(f"Page {page_num + 1}, Image {img_index + 1}:\n{text}\n")


ocr_pdf("../Data/PDF/scan/scan1.pdf")
```

```python
import easyocr  # type: ignore
from pdf2image import convert_from_path # type: ignore
from PIL import Image  # type: ignore

def pdf_to_images(pdf_path, dpi=900):
    images = convert_from_path(pdf_path, dpi)
    return images

def ocr_from_images(images, lang='ru'):
    reader = easyocr.Reader([lang])
    full_text = []
    for img in images:
        result = reader.readtext(img, paragraph=True)
        full_text.append('\n'.join(result))
    return '\n\n'.join(full_text)

def ocr_from_pdf(pdf_path, lang='ru', output_file=None):
    images = pdf_to_images(pdf_path)
    recognized_text = ocr_from_images(images, lang)

    if output_file:
        with open(output_file, 'w', encoding='utf-8') as f:
            f.write(recognized_text)
        print('done')
    else:
        print(recognized_text)

path = 'Data\PDF\scan\scan1.pdf'
ocr_from_pdf(path, lang='ru')
```
